```python
datos["FECHA DE RADICACION"] = pd.to_datetime(datos["FECHA DE RADICACION"])
datos["FECHA RADICACION DE RESPUESTA"] = pd.to_datetime(datos["FECHA RADICACION DE RESPUESTA"], errors="coerce")
```

```python
datos["TIEMPO DE RESPUESTA"] = datos["FECHA RADICACION DE RESPUESTA"] - datos["FECHA DE RADICACION"]

datos.head()
```

```python
datos["TIEMPO DE RESPUESTA"].mean()
```
